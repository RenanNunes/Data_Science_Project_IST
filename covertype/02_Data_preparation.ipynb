## Normalization

The variables have different scales, which can influentiate the models, that's why we will normalize the data.

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import pandas as pd
import numpy as np
from utils import dataPreparationFunctions as dataPrepare

data = pd.read_csv('../datasets/covtype_data_exploration.csv', sep=',', decimal='.')
data.head()
```

```python
norm_data = dataPrepare.normalize_data(data.iloc[:, 0:data.shape[1]-1])
norm_data = pd.concat([norm_data, data.iloc[:, data.shape[1]-1]], axis=1)
norm_data.head()
```

## Data balancing

We have an unbalanced dataset, therefore we can use undersampling to get a smaller amount of data to train the models, because of the high number of records.

```python
norm_data['Cover_Type'].value_counts()
```

```python
prepared_data = dataPrepare.undersample(norm_data, 'Cover_Type')
prepared_data.columns = data.columns
prepared_data['Cover_Type'].value_counts()
```

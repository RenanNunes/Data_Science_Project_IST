# Lab 7 - Clustering

We will create cluster (groups of records/patients) according to their speech features and analyze it. For this, we are going to use K-means with differente values of k to see the impact of the parameter.

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import pandas as pd
import numpy as np
from sklearn import metrics, cluster
import matplotlib.pyplot as plt
```

```python
data_9: pd.DataFrame = pd.read_csv('../datasets/pd_data_preparation_9.csv', sep=',', decimal='.', index_col='id')
y_9: np.ndarray = data_9.pop('class').values
X_9: np.ndarray = data_9.values
labels_9 = pd.unique(y_9)

data_9.head()
```

```python
n_clusters = [2, 3, 5, 8, 10, 15, 20, 30, 40, 50, 75, 100]
results = {}
for k in n_clusters:
    results[k] = {}
for k in n_clusters:
    kmeans = cluster.KMeans(n_clusters=k, random_state=1).fit(X_9)
    y_pred = kmeans.labels_
    value_counts = pd.Series(y_pred).value_counts()
    results[k] = {
        "sse": kmeans.inertia_,
        "silhouette": metrics.silhouette_score(X_9, y_pred),
        "small_clusters": value_counts[value_counts == 1].shape[0]
    }

# fig = plt.figure()
# ax = plt.gca()
# ax.set_title("Sum of squared distances by number of clusters")
# ax.plot(results.keys(), results.sse.values())
# plt.show()

# fig = plt.figure()
# ax = plt.gca()
# ax.set_title("Silhouette by number of clusters")
# ax.plot(silhouette.keys(), silhouette.values())
# plt.show()
```

The high number of outliers and features by the low number of features result in clusters not that well divided, represented by the metric Sum of Squared Distances. One way to minimize the distance would be increase the number of clusters, but it can create a lot of clusters with low number of records, which may not be so useful.
